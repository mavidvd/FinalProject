### Joining other data.
For the purpose of our research, we are interested in checking whether there is any correlation between some demographic and economic variables relative to each city or state and the amount of flights operating in such state. 

Our next steps are going to be obtaining the data for the state and city information we are willing to work with and join it to the main dataframe.

```python
stdf = pd.read_excel('Data/StateData.xlsx')
```

```python
stdf
```

```python
stdf.columns = ['State', 'ST', 'Pop', 'GDP', 'Share', 'GDPpc']
```

```python
stdf.drop(['Pop'], axis = 1, inplace = True)
stdf.drop(['GDP'], axis = 1, inplace = True)
```

```python
stdf['Share'] = stdf['Share']*100
```

```python
stdf[['Share', 'GDPpc']] = round(stdf[['Share', 'GDPpc']], 2)
```

```python
df = df.merge(stdf, left_on = 'FST', right_on = 'ST', how = 'left')
df = df.merge(stdf, left_on = 'TST', right_on = 'ST', how = 'left')
```

```python
df
```

```python
df.columns
```

```python
df.columns = ['Route', 'Month', 'Carrier', 'From', 'FCity', 'FST', 'To', 'TCity', 'TST', 
              'Delay', 'Flights', 'Dist', 'State_x', 'ST_x', 'FShare', 'FGDPpc',
              'State_y', 'ST_y', 'TShare', 'TGDPpc']
```

```python
df = df[['Route', 'Month', 'Carrier',
          'From', 'FCity', 'FST', 'FShare', 'FGDPpc',
          'To', 'TCity', 'TST', 'TShare', 'TGDPpc',
          'Delay', 'Flights', 'Dist']]
```

```python
df
```
